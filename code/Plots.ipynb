# Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Plot metrics

## Aggregation = Mean

```python
df_mean = pd.read_csv(r'C:\Users\franc\Downloads\metrics_mean (2).csv')
df_mean.head()
```

```python
plt.figure(figsize=(8,5), facecolor='white')

# Plotting loss
plt.plot(df_mean["epoch"], df_mean["loss"], label="Train Loss", color="red", linewidth=2)

plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training loss over epochs (agg=mean)")
plt.grid(True, color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.gca().set_facecolor('white')
plt.tight_layout()

plt.savefig("training_loss_mean.png", dpi=300, bbox_inches='tight')  # Save as PNG with high resolution

plt.show()
```

```python
plt.figure(figsize=(8,5), facecolor='white')

plt.plot(df_mean["epoch"], df_mean["train_acc"] * 100, label="Train Accuracy", color="green", linewidth=2)

plt.xlabel("Epoch")
plt.ylabel("Accuracy (%)")
plt.title("Training Accuracy over Epochs (agg=mean)")
plt.grid(True, color='gray', linestyle='--', linewidth=0.5)
plt.ylim(0, 60)
plt.legend()
plt.tight_layout()

plt.savefig("training_acc_mean.png", dpi=300, bbox_inches='tight')  # Save as PNG with high resolution

plt.show()
```

# Venue Distribution
